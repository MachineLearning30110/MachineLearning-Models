```python
from sklearn.datasets import load_diabetes
import pandas as pd

data = load_diabetes()

df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

print(df.head())
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42
)

print(X_test.shape)
```

**Ridge Regularization (L2)**

    Loss + λ ∑ w^2

*   Shrinks weights
*   Keeps all features



```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

ridge_preds = ridge.predict(X_test)
```

```python
from sklearn.linear_model import RidgeCV

ridgeCV = RidgeCV(alphas=(0.1,1,10))
ridgeCV.fit(X_train, y_train)

ridgeCV_preds = ridgeCV.predict(X_test)
```

**Lasso Regularization (L1)**

    Loss + λ ∑|w|
    
*   Drives some weights to zero
*   Performs feature selection



```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

lasso_preds = lasso.predict(X_test)
```

```python
print(lasso.coef_)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

print("Linear:", r2_score(y_test, model.predict(X_test)))
print("Ridge:", r2_score(y_test, ridge_preds))
print("RidgeCV:", r2_score(y_test, ridgeCV_preds)," ,alpha =",ridgeCV.alpha_)
print("Lasso:", r2_score(y_test, lasso_preds))
```
